```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ','_')
)
df.columns
```

```python
df.columns = (
    df.columns
      .str.replace(r'[^A-Za-z0-9]+', '_', regex=True)
      .str.strip('_')
      .str.lower()
)

df.columns
```

```python
df.columns = [
    'business_category',
    'business_name',
    'business_contact',
    'business_location',
    'mobile_contact',
    'office_contact'
]

df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['office_contact'])
df['primary_contact'] = df['mobile_contact'].fillna(df['business_contact'])
df = df.dropna(subset=['primary_contact'])
```

```python
df_selected = df[
    df['business_category'].str.contains(
        'restaurant|food|cafe|eatery|salon|beauty|hairstylist|barber',
        case=False,
        na=False
    )
]
```

```python
df_selected = df_selected.reset_index(drop=True)
```

```python
df_selected['business_category'].value_counts()
```

```python
# Number of SMEs you want in final dataset
final_sample_size = 30

# Calculate proportional sampling
category_counts = df_selected['business_category'].value_counts()
category_proportions = category_counts / category_counts.sum()

# Sample from each category proportionally
df_final = df_selected.groupby('business_category', group_keys=False).apply(
    lambda x: x.sample(max(1, round(len(x) / len(df_selected) * final_sample_size)))
)

# Reset index
df_final = df_final.reset_index(drop=True)
```

```python
df_final.to_excel("sme_interview_final_dataset.xlsx", index=False)
```
